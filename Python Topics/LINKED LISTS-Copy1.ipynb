```python
# Creating LinkedList and Inserting and Printing values in the LinkedList
class LinkedList:
    def __init__(self):
        self.head=None
    def insert(self,data):
        a=Node(data)
        if(self.head is None):
            self.head=a
        else:
            b=self.head
            while b.next:
                b=b.next
            b.next=a
    def view(self):
        b=self.head
        while b:
            print(b.data,end="-->")
            b=b.next
```

```python
# Merge of two LinkedLists
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class LinkedList:
    def __init__(self):
        self.head=None
    def insert(self,value):
        j=Node(value)
        if self.head is None:
            self.head=j
        else:
            q=self.head
            while q.next:
                q=q.next
            q.next=j
    def view(self):
        q=self.head
        while q:
            print(q.data,end=" ")
            q=q.next
def merge_of_elements(head1,head2):
    dummy=Node(0)
    tail=dummy
    while True:
        if head1 is None:
            tail.next=head2
            break
        if head2 is None:
            tail.next=head1
            break
        if head1.data<=head2.data:
            tail.next=head1
            head1=head1.next
        else:
            tail.next=head2
            head2=head2.next
        tail=tail.next
    return dummy.next
list1=LinkedList()
list2=LinkedList()
for i in range(5):
    list1.insert(int(input()))
for i in range(5):
    list2.insert(int(input()))
print("List one Elements")
list1.view()
print()
print("List two Elements")
list2.view()
print()
list1.head = merge_of_elements(list1.head, list2.head)
print("Merged Linked List is:")
list1.view()
```
